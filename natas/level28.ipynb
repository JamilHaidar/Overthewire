```python
import string
def test_query(query):
    data = {'query': query}
    r = session.post(url, data=data)
    decoded_url = unquote(r.url.split('query=')[1])
    blocks = base64.b64decode(decoded_url).hex()
    blocks = [blocks[i:i+32] for i in range(0, len(blocks), 32)]
    # for i,elem in enumerate(blocks):
    #     print(i,elem)
    return decoded_url,blocks

def print_url(decoded_url):
    url = [decoded_url[i:i+16] for i in range(0, len(decoded_url), 16)]
    print(url)


initial_decoded_url,initial_blocks = test_query('A')
test_decoded_url,test_blocks = test_query('A'+'%'*9)
print(initial_decoded_url,initial_blocks)
print(test_decoded_url,test_blocks)
found_padding = 'A'*9
block_count = 2
for _ in range(9):
    print('Finding',found_padding)
    initial_decoded_url,initial_blocks = test_query(found_padding)
    for test_char in '%'+string.printable:
        test_decoded_url,test_blocks = test_query(found_padding+test_char)
        if initial_blocks[block_count] == test_blocks[block_count]:
            print('Found',test_char)
            print(initial_blocks[block_count],test_blocks[block_count])
            print_url(initial_decoded_url)
            print_url(test_decoded_url)
            found_padding =  found_padding[1:] + test_char
            print(found_padding)
            break
```
